```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marvel = pd.read_csv('marvel_powers.csv')
```

```python
marvel.head()
```

#### Get frequency of types of super heroes

```python
marvel['Alignment_x'].value_counts()
```

```python
for col in marvel.columns:
    print(col)
```

```python
good_marvel = marvel[marvel['Alignment_x'] == 'good']
```

```python
good_marvel.head()
```

### Create a count plot of all character types, categorize based on gender

```python
sns.countplot(data=marvel, x='Alignment_x', hue='Gender')
```

```python
# Fastest 5 good characters
good_marvel[['Name','Gender','Race','Speed']].sort_values(by='Speed', ascending=False).head()
```

```python
# Powerful top5 characters
good_marvel[['Name','Gender','Race','Speed','Power']].sort_values(by='Power', ascending=False).head()
```

```python
# Most powerful good character in terms of total points
good_marvel[['Name', 'Gender','Total']].sort_values(by='Total',ascending=False).head()
```

### Get all bad characters

```python
bad_marvel = marvel[marvel['Alignment_x'] == 'bad']
```

```python
bad_marvel.head()
```

```python
# 5 most powerful bad characters
bad_marvel[['Name','Power','Gender','Speed']].sort_values(by='Power', ascending=False).head()
```

```python
# 5 most fastest characters
bad_marvel[['Name','Power','Gender','Speed']].sort_values(by='Speed', ascending=False).head()
```

#### count plot of bad marvels based on gender

```python
sns.countplot(data=bad_marvel, x='Gender', hue='Race')
```

```python
# Most powerful in terms of total character in bad 
bad_marvel[['Name', 'Gender', 'Total']].sort_values(by='Total', ascending=False).head()
```
